```sql
-- Example SQL query to extract incident data from the database
SELECT incident_id, incident_type, severity, date_reported, status
FROM incidents
WHERE date_reported BETWEEN '2023-01-01' AND '2023-12-31'
ORDER BY severity DESC;

## 2. Loading Data into Python

The following Python code connects to a SQL Server, executes a query to fetch cybersecurity incident data, and loads it into a Pandas DataFrame for analysis.

```python
import pandas as pd
import pyodbc

# Connect to SQL Server
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=your_server;DATABASE=cybersecurity;UID=user;PWD=password')

# Query the data
query = '''
SELECT 
    IncidentID,
    IncidentType,
    SeverityLevel,
    AffectedSystem,
    ResolutionTime,
    Timestamp,
    Geolocation
FROM 
    CyberSecurityIncidents
WHERE 
    Timestamp BETWEEN '2023-01-01' AND '2023-12-31';
'''

# Load data into a Pandas DataFrame
df =
(query, conn)
df.head()
query, conn)
df.head()


## 3. Exploratory Data Analysis (EDA)

### Incident Distribution by Severity

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count of incidents by severity level
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='SeverityLevel', palette='coolwarm')
plt.title('Incident Distribution by Severity')
plt.xlabel('Severity Level')
plt.ylabel('Count of Incidents')
plt.show()


## Frequency of Incidents Over Time

```python
# Converting Timestamp to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Grouping by date to calculate incident frequency
incident_freq = df.groupby(df['Timestamp'].dt.date).size()

# Plotting
plt.figure(figsize=(10,6))
incident_freq.plot(kind='line', color='b')
plt.title('Cybersecurity Incidents Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Incidents')
plt.grid(True)
plt.show()


## 4. Key Metrics and Insights

### Average Resolution Time by Severity Level

```python
# Calculating mean resolution time by severity level
resolution_by_severity = df.groupby('SeverityLevel')['ResolutionTime'].mean()

# Plotting
plt.figure(figsize=(8,5))
resolution_by_severity.plot(kind='bar', color='teal')
plt.title('Average Resolution Time by Severity Level')
plt.xlabel('Severity Level')
plt.ylabel('Average Resolution Time (hours)')
plt.show()


### Geolocation of Incidents

```python
# Grouping by geolocation and counting incidents
geo_incidents = df.groupby('Geolocation').size()

# Plotting as a horizontal bar chart
plt.figure(figsize=(10,8))
geo_incidents.plot(kind='barh', color='orange')
plt.title('Cybersecurity Incidents by Geolocation')
plt.xlabel('Number of Incidents')
plt.ylabel('Geolocation')
plt.show()


## 5. Interactive Dashboard with Filters (Optional)

For the dashboard, you can use **Tableau** or **Power BI** to import the CSV dataset and create dynamic visualizations with filters. Here's a basic concept for a dashboard:

- **Overview**: Total number of incidents, most frequent incident type.
- **Incident Trends**: Time-series analysis of incidents.
- **Geolocation Map**: Interactive map of incidents based on geolocation.
- **Resolution Time**: Distribution of resolution times for each severity.

## 6. Conclusion and Recommendations

### Critical Insights:
- Most incidents are concentrated in high-severity levels, affecting critical systems.
- Incidents are growing in certain geolocations (highlighted in the map).
- Resolution times increase as severity level rises.

### Action Items:
- Strengthen security measures in regions with the highest incident rates.
- Investigate systems that are frequently affected.
- Optimize response protocols for high-severity incidents to reduce resolution time.


## Frequency of Incidents Over Time

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Count of Incidents': [5, 3, 8, 4, 6]
}

# Creating a DataFrame
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Count of Incidents'], marker='o', linestyle='-', color='b')
plt.title('Frequency of Incidents Over Time')
plt.xlabel('Date')
plt.ylabel('Count of Incidents')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
