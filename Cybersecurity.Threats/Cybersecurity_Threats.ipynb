````python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Timestamp': ['2024-01-01 10:00:00', '2024-01-02 11:00:00', '2024-01-03 12:00:00',
                      '2024-01-04 13:00:00', '2024-01-05 14:00:00'],
        'SeverityLevel': ['High', 'Medium', 'Low', 'High', 'Medium']}

# Creating a DataFrame
df = pd.DataFrame(data)

# Ensure the Timestamp column is in datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Grouping by date to calculate incident frequency
incident_freq = df.groupby(df['Timestamp'].dt.date).size()

# Plotting the frequency of incidents over time
plt.figure(figsize=(10, 6))
incident_freq.plot(kind='line', marker='o', color='purple')
plt.title('Frequency of Incidents Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Number of Incidents', fontsize=12)
plt.grid(True)
plt.show()
````

## 4. Key Metrics and Insights

### Average Resolution Time by Severity Level

```python
# Calculating mean resolution time by severity level
resolution_by_severity = df.groupby('SeverityLevel')['ResolutionTime'].mean()

# Plotting
plt.figure(figsize=(8,5))
resolution_by_severity.plot(kind='bar', color='teal')
plt.title('Average Resolution Time by Severity Level')
plt.xlabel('Severity Level')
plt.ylabel('Average Resolution Time (hours)')
plt.show()


````python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (adding a ResolutionTime column)
data = {'Timestamp': ['2024-01-01 10:00:00', '2024-01-02 11:00:00', '2024-01-03 12:00:00',
                      '2024-01-04 13:00:00', '2024-01-05 14:00:00'],
        'SeverityLevel': ['High', 'Medium', 'Low', 'High', 'Medium'],
        'ResolutionTime': [5, 3, 2, 6, 4]}  # Adding ResolutionTime in hours

# Creating a DataFrame
df = pd.DataFrame(data)

# Ensure the Timestamp column is in datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Calculating mean resolution time by severity level
resolution_by_severity = df.groupby('SeverityLevel')['ResolutionTime'].mean()

# Plotting Average Resolution Time by Severity Level
plt.figure(figsize=(8, 5))
resolution_by_severity.plot(kind='bar', color='skyblue')
plt.title('Average Resolution Time by Severity Level', fontsize=14)
plt.xlabel('Severity Level', fontsize=12)
plt.ylabel('Average Resolution Time (hours)', fontsize=12)
plt.xticks(rotation=0)
plt.show()
````

### Geolocation of Incidents

```python
# Grouping by geolocation and counting incidents
geo_incidents = df.groupby('Geolocation').size()

````python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data with Geolocation column
data = {'Timestamp': ['2024-01-01 10:00:00', '2024-01-02 11:00:00', '2024-01-03 12:00:00',
                      '2024-01-04 13:00:00', '2024-01-05 14:00:00'],
        'SeverityLevel': ['High', 'Medium', 'Low', 'High', 'Medium'],
        'Geolocation': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Los Angeles']}

# Creating a DataFrame
df = pd.DataFrame(data)

# Ensure the Timestamp column is in datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Grouping by geolocation and counting incidents
geo_incidents = df.groupby('Geolocation').size()

# Plotting the Geolocation of Incidents
plt.figure(figsize=(10, 8))
geo_incidents.plot(kind='bar', color='coral')
plt.title('Geolocation of Incidents', fontsize=14)
plt.xlabel('Geolocation', fontsize=12)
plt.ylabel('Number of Incidents', fontsize=12)
plt.xticks(rotation=45)
plt.show()
````
